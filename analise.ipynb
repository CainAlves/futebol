Analisando dados de Futebol

```python
# Importando bibliotecas necessarias 
import basedosdados as bd
import pandas as pd 

# Para carregar o dado direto no pandas
df = bd.read_table(dataset_id='mundo_transfermarkt_competicoes',
table_id='brasileirao_serie_a',
billing_project_id="futebol-394512")
df.head()
```

```python
# Analisando a tabela extraida do site basedosdados.org 
resumo = df.describe()
print(resumo)
print(df.columns)
```

```python
# Limpando a base para analisar somente o ano corrente 
df_2023 = df[df['ano_campeonato'] == 2023]
df_2023.head()
```

Quais times possues as maiores médias de público ? 

```python
# Pegando a media de publico presente dos times mandantes 
media_publico = df_2023.groupby('time_man')['publico'].mean()
media_publico = media_publico.round(1)
media_publico = media_publico.sort_values(ascending=False)
print(media_publico)
```

Quais times fizeram mais gols dentro de casa? (Melhores ataques mandantes)

```python
# Somando a quantidade de gols dos times mandante 
gols = df_2023.groupby('time_man')['gols_man'].sum()
gols = gols.sort_values(ascending=False)
print(gols)
```

Quais são os times mais violentos? (com mais faltas)

```python
# Separando as faltas por time enquanto mandante e visitando para calcular o total de faltas de cada um 
df_mandantes = df_2023.groupby('time_man')['faltas_man'].sum()
df_visitantes = df_2023.groupby('time_vis')['faltas_vis'].sum()

df_mandantes = pd.DataFrame(df_mandantes)
df_mandantes.reset_index(level=0, inplace=True)
df_visitantes = pd.DataFrame(df_visitantes)
df_visitantes.reset_index(level=0, inplace=True)

df_final = pd.merge(df_mandantes, df_visitantes, left_on='time_man', right_on='time_vis', how='left')
df_final['faltas_total'] = df_final['faltas_man'] + df_final['faltas_vis']
df_final = df_final[['time_man','faltas_total']].rename(columns={'time_man': 'time'})
print(df_final)

```
